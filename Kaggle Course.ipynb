### Picking some  variables

```python
import pandas as pd

# file path location
melbourne_file_path = 'input/CSV/melb_data.csv'
# store the data in a DataFrame named melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path)
melbourne_data.columns
melbourne_data
```

Drops the lines where there is missing values

```python
melbourne_data.dropna(axis=0)
```

### Choose a target
Dot-notation: choose the column you want to predict. This column is the target, by convention stored in y.

```python
y = melbourne_data.Price
```

### Chose the features
Those columns that will be used as input to the modle (and latter used to predict the target) are called features.

In this case, the columns used to predict the **price** are the targets.

**We select multiple features by providing a list of column names inside brackets.**

```python
melbourne_features = ["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]
```

By convention this data is called X:

```python
X = melbourne_data[melbourne_features]
X.describe()
```

## Buiild a Model

The steps to building and using a model are:

- Define: What type of model will it be? A decision tree? Some other type of model? Some other parameters of the model type are specified too.
- Fit: Capture patterns from provided data. This is the heart of modeling.
- Predict: Just what it sounds like
- Evaluate: Determine how accurate the model's predictions are.


```python
from sklearn.tree import DecisionTreeRegressor

# Define a model (random_state=1 ensures same results every run)
melbourne_model = DecisionTreeRegressor(random_state=1)

# Train the model
melbourne_model.fit(X, y)
```

## Predicting

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```
